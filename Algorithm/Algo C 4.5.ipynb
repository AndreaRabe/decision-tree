# Dataset separation test

```python
dataset_golf = pd.read_csv('/content/drive/MyDrive/Skills/Python/Data mining/Decision Tree/End of year project 2024/Datasets/golf.txt')
dataset_golf
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(dataset_golf, test_size=0.4, random_state=42)
```

## Scikit-learn

```python
X = dataset_golf[['Outlook', 'Temp.', 'Humidity', 'Wind']]
y = dataset_golf[['Decision']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 52)
```

## Let's train

```python
config = {'algorithm' : 'C4.5'}
reg_model = chef.fit(train_set.copy(), config)
```

```python
for index, instance in test_set.iterrows():
  print(instance , "Prediction: ", chef.predict(reg_model, instance))
  # print(instance)
```

```python
# Ajoutez une colonne pour stocker les prédictions
test_set['prediction'] = test_set.apply(lambda instance: chef.predict(reg_model, instance), axis=1)

# Comparez les prédictions avec les vraies valeurs
correct_predictions = (test_set['prediction'] == test_set['Decision']).sum()  # Remplacez 'target_column' par le nom de la colonne cible

# Calculez l'accuracy
accuracy = correct_predictions / len(test_set)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
