# Importation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[1, 2], [1, 2.5], [2, 1], [3, 1.5], [2, 2.5], [3.5, 1], [4, 2], [3, 3]])
y = [0, 0, 0, 0, 1, 1, 1, 1]

# X = np.array([[0,0], [1,0], [2,0], [3,0], [0.5,1], [1.5,1], [2.5,1], [3.5,1]])
# y = [0, 0, 0, 0, 1, 1, 1, 1]
```

```python
plt.scatter(X[:,0],X[:,1], c=y)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier() # Gini par defaut si on ne precise pas
clf.fit(X, y)
```

```python
max_depth = clf.get_depth()
print(max_depth)
```

```python
from sklearn import tree
tree.plot_tree(clf, rounded=True, filled=True)
plt.show()
```

# With Entropy

```python
clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(X, y)
```

```python
max_depth = clf.get_depth()
print(max_depth)
```

```python
tree.plot_tree(clf, rounded=True, filled=True)
plt.show()
```
